```python
# Aditional Exercise no.1 

import time

# Inputs
student_id = "1000-5000"
main_list = list(range(1000))  # Fixed list for search
duplicate_list = [10, 22, 14, 33, 22, 5, 8]  # List with a duplicate

# Calculations
target = int(student_id[-2:])  # Target from last two digits

# Linear Search (O(n))
start = time.perf_counter()
for i in main_list:
    if i == target: break
linear_time = time.perf_counter() - start

# Binary Search (O(log n))
start = time.perf_counter()
low, high = 0, len(main_list) - 1
while low <= high:
    mid = (low + high) // 2
    if main_list[mid] == target: break
    elif main_list[mid] < target: low = mid + 1
    else: high = mid - 1
binary_time = time.perf_counter() - start

# Quadratic Algorithm (O(n^2))
start = time.perf_counter()
for i in range(len(duplicate_list)):
    for j in range(i + 1, len(duplicate_list)):
        if duplicate_list[i] == duplicate_list[j]: break
quadratic_time = time.perf_counter() - start

# Output
print(f"{'Algorithm':<20} | {'Time Taken (s)':<15}")
print("-" * 40)
print(f"{'Linear Search':<20} | {linear_time:.8f}")
print(f"{'Binary Search':<20} | {binary_time:.8f}")
print(f"{'Nested Loop':<20} | {quadratic_time:.8f}")

print("\nAnalysis: Binary search (logarithmic) scales best, while the nested loop (quadratic) "
      "will see the most significant time increase as data grows.")
```

```python
# Aditional Exercise no.2


import time

# Input
student_id = "1000-5000"

# Calculations
n = int(student_id[-1]) + 15  # Last digit + 15
print(f"Calculated Problem Size (n): {n}")

start = time.perf_counter()
# Iterative Fibonacci Algorithm
a, b = 0, 1
for _ in range(n):
    a, b = b, a + b
result = a
execution_time = time.perf_counter() - start

# Output
print(f"\n{'Problem Size (n)':<18} | {'Time Taken (s)':<15} | {'Result':<10}")
print("-" * 50)
print(f"{n:<18} | {execution_time:.8f} | {result:<10}")

print("\nDiscussion: The iterative Fibonacci algorithm has a linear Big O complexity, O(n), "
      "as it only traverses the loop once for each value up to n.")
```

```python
# Aditional Exercise no.3

import time
import random

# Input
student_id = "1000-5000"
list_size = int(student_id[-3:]) + 50 # Last 3 digits + 50
test_data = random.sample(range(1, 2000), list_size)

def bubble_sort(arr):
    comp = swaps = 0
    n = len(arr)
    for i in range(n):
        for j in range(0, n-i-1):
            comp += 1
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
                swaps += 1
    return comp, swaps

def selection_sort(arr):
    comp = swaps = 0
    n = len(arr)
    for i in range(n):
        min_idx = i
        for j in range(i+1, n):
            comp += 1
            if arr[j] < arr[min_idx]: min_idx = j
        arr[i], arr[min_idx] = arr[min_idx], arr[i]
        swaps += 1
    return comp, swaps

def insertion_sort(arr):
    comp = swaps = 0
    for i in range(1, len(arr)):
        key = arr[i]
        j = i-1
        while j >= 0:
            comp += 1
            if key < arr[j]:
                arr[j+1] = arr[j]
                swaps += 1
                j -= 1
            else: break
        arr[j+1] = key
    return comp, swaps

# Execution and Output
print(f"Sorting list of size: {list_size}\n")
print(f"{'Algorithm':<15} | {'Comparisons':<12} | {'Swaps':<8} | {'Time (s)':<10}")
print("-" * 55)

for name, func in [("Bubble", bubble_sort), ("Selection", selection_sort), ("Insertion", insertion_sort)]:
    data_copy = test_data.copy()
    start = time.perf_counter()
    c, s = func(data_copy)
    duration = time.perf_counter() - start
    print(f"{name:<15} | {c:<12} | {s:<8} | {duration:.5f}")

print("\nRecommendation: Insertion sort is typically most efficient for smaller or partially sorted sets, "
      "while Selection sort minimizes the total number of swaps.")
```
